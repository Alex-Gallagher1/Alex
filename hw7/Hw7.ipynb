```python
#File: Hw7.ipynb
import numpy as np
import matplotlib.pyplot as plt
```

```python
#---2 Matplotlib---
def make_sine_wave(x,A,w):
    y = A*np.sin(w*x)
    return y

x = np.linspace(0,2*np.pi,1000)
amplitude =  np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
linestyles = np.array(["-","--","-.",":","-"])
```

```python
fig, axs = plt.subplots(1,3,figsize=(25,5))
params = zip(amplitude, frequency, linestyles)

for i,j,k in params:
    y_amplitude = make_sine_wave(x,i,1)
    y_freq = make_sine_wave(x,1,j)
    y_both = make_sine_wave(x,i,j)
    axs[0].plot(x,y_amplitude, linestyle = k, label =f'A={i} w = {1}')
    axs[1].plot(x,y_freq, linestyle = k, label =f'A={1} w = {j}')
    axs[2].plot(x,y_both, linestyle = k, label =f'A={i} w = {j}')
    axs[0].legend()
    axs[1].legend()
    axs[2].legend()

plt.suptitle("Affects of Amplitude and Frequency Modulation on Sine Wave", fontsize = 15, )
axs[0].set_title("Amplitude Modulation")
axs[1].set_title("Frequency Modulation")
axs[2].set_title("Amplitude and Frequency Modulation")
axs[0].set_xlabel("x", fontsize = 15)
axs[1].set_xlabel("x", fontsize = 15)
axs[2].set_xlabel("x", fontsize = 15)
axs[0].set_ylabel("y", fontsize = 15)
axs[1].set_ylabel("y", fontsize = 15)
axs[2].set_ylabel("y", fontsize = 15)

plt.show()
```
