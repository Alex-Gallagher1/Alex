```python
#File: hr_diagram.ipynb
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("star_data.txt", skiprows=1)
```

```python
np.shape(data) #There are 2642 rows and 2 columns
magnitude = data[:,0]
temperature = data[:,1]

min_mag = min(magnitude)
max_mag = max(magnitude)

min_temp = min(temperature)
max_temp = max(temperature)
print(min_mag, max_mag, min_temp, max_temp)

min_mag_index = np.argmin(magnitude) #These next few variables define the index of the minimum and maximum values of magnitude and temperature
max_mag_index = np.argmax(magnitude)

min_temp_index = np.argmin(temperature)
max_temp_index = np.argmax(temperature)

brightest_star = data[1988]
dimmest_star = data[264]
hottest_star = data[2613]
coldest_star = data[1923]
print(min_mag_index, max_mag_index, min_temp_index, max_temp_index) #the brightest star is the star at index 1988 and the dimmest star is at index 264
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize = (8,6))
plt.scatter(log_temperature, log_magnitude,c=log_temperature, cmap = "plasma", s = 5, label = "Stars")


#special stars
plt.scatter(np.log(brightest_star[1]), np.log(brightest_star[0]), c = "black", marker = "*", s = 300, label = "Brightest Star")
plt.scatter(np.log(dimmest_star[1]), np.log(dimmest_star[0]), c = "purple", marker = "*", s = 300, label = "Dimmest Star")
plt.scatter(np.log(hottest_star[1]), np.log(hottest_star[0]), c = "green", marker = "*", s = 300, label = "Hottest Star")
plt.scatter(np.log(coldest_star[1]), np.log(coldest_star[0]), c = "orange", marker = "*", s = 300, label = "Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.legend()

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.show()
```
